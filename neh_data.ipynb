```python
gender_discipline = [
    "GENDER STUDIES",
    "WOMEN'S HISTORY"
]
```

```python
info_columns = ['ProjectTitle', 'ProjectDesc', 'ToSupport']
discipline_columns = ['PrimaryDiscipline', 'Disciplines']
```

```python
# Below is my original code to establish a gender_filter.
# It didn't work because pandas needed to evaluate it as a string, which is was not.

# def gender_filter(row):
#    for column in info_columns:
#        if any(keyword in row[column].upper() for keyword in gender_keywords):
#            return True
#    for column in discipline_columns:
#        if any(keyword in row[column].upper() for keyword in gender_discipline):
#            return True
#    return False

# Below, the code is revised to convert each row[column] to a string before applying the upper() method.

def gender_filter(row):
    for column in info_columns:
        if any(keyword in str(row[column]).upper() for keyword in gender_keywords):
            return True
    for column in discipline_columns:
        if any(keyword in str(row[column]).upper() for keyword in gender_discipline):
            return True
    return False
```

```python
combined_df['gender_focus'] = combined_df.apply(gender_filter, axis=1)

# apply lets me run the function across rows or columns
# axis=1 tells pandas to specifically apply the function to each row of the dataframe
```

```python
# This line creates a table showing projects with/without a gender_focus by year.

gender_focus_by_year = combined_df.groupby(["YearAwarded", "gender_focus"]).size().rename("projects").reset_index()
```

```python
# Using "pivot", this code creates a pivot table by identifing the index (row) as gender_focus,
# the columns as YearAwarded,
# and the values inside them as the projects.

# "fillna" means that any empty cells without a number (NaN - "not a number")
# will be filled with 0 instead of NaN.
# So if there are no gender-related projects in a year, Python will put a 0 instead of NaN.

gender_year_pivot_table = gender_focus_by_year.pivot(
    index="YearAwarded",
    columns="gender_focus",
    values="projects"
).fillna(0)
```

```python
gender_year_pivot_table

# This could be good to visualize in a stacked bar chart
# with a line graph overlay, showing the rate of change.
```

```python
# This line creates a table showing projects with/without a gender_focus by state.

gender_focus_by_state = combined_df.groupby(["InstState", "gender_focus"]).size().rename("projects").reset_index()
```

```python
gender_state_pivot_table = gender_focus_by_state.pivot(
    index="InstState",
    columns="gender_focus",
    values="projects"
).fillna(0)
```

```python
# aggfunc means "aggregation function", which tells pandas how to summarize the data.
# Here, I use "sum" to add up all the AwardOutright amounts for each year for projects with and without a gender_focus.
# fill_value=0, like fillna from before, ensures that any empty cells are filled with 0 instead of NaN.

award_per_year_pivot = combined_df.pivot_table(
    index="YearAwarded",
    columns="gender_focus",
    values="AwardOutright",
    aggfunc="sum",
    fill_value=0
)
```
